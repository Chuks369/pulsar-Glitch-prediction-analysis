```python
import numpy as np
import hdbscan
import pandas as pd

# Load your data
df = pd.read_excel('\\Users\\hp\\Desktop\\MALACHY.xlsx')

# Extract MJD and ensure it's a writable, contiguous float64 array
mjds_raw = df["MJD"].values

# Convert to float64 and make it C-contiguous (writable)
mjds = np.ascontiguousarray(mjds_raw.astype(np.float64)).reshape(-1, 1)

# Verify the data
print(f"Shape: {mjds.shape}")
print(f"Dtype: {mjds.dtype}")
print(f"Contiguous: {mjds.flags['C_CONTIGUOUS']}")
print(f"Writable: {mjds.flags['WRITEABLE']}")
print(f"First 5 MJD values:\n{mjds[:5]}")

# Run HDBSCAN with a copy to avoid any readonly issues
clusterer = hdbscan.HDBSCAN(min_cluster_size=2)
labels = clusterer.fit_predict(mjds.copy())  # Use .copy() to ensure writable

print(f"\nCluster labels: {labels}")
print(f"Number of clusters found: {len(set(labels)) - (1 if -1 in labels else 0)}")
print(f"Noise points (labeled -1): {sum(labels == -1)}")
```
