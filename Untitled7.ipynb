```python
Distribution Fitting (AIC/BIC)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import expon, norm

# Load the data
df = pd.read_excel('\\Users\\hp\\Desktop\\MALACHY.xlsx')

# Calculate waiting times (differences between consecutive MJDs)
# Sort by MJD first
df_sorted = df.sort_values('MJD')
waiting_times = np.diff(df_sorted['MJD'].values)

# Remove any zero or negative waiting times if they exist
waiting_times = waiting_times[waiting_times > 0]

print(f"Calculated {len(waiting_times)} waiting times between glitches")
print(f"Waiting times (days): {waiting_times}")

# Fit distributions to waiting times
params_exp = expon.fit(waiting_times)
params_norm = norm.fit(waiting_times)

# Log-likelihoods
ll_exp = np.sum(expon.logpdf(waiting_times, *params_exp))
ll_norm = np.sum(norm.logpdf(waiting_times, *params_norm))

# AIC
aic_exp = 2*len(params_exp) - 2*ll_exp
aic_norm = 2*len(params_norm) - 2*ll_norm

print("\n--- Distribution Fitting Results ---")
print(f"AIC (Exponential): {aic_exp:.2f}")
print(f"AIC (Gaussian): {aic_norm:.2f}")

if aic_exp < aic_norm:
    print("\n✓ Exponential distribution fits better (glitches follow Poisson process)")
else:
    print("\n✓ Gaussian distribution fits better")
```

```python

# Step 1: Load the data
df = pd.read_excel('\\Users\\hp\\Desktop\\MALACHY.xlsx')

# Step 2: Create the Cluster column using HDBSCAN
# Use MJD as feature
X = df[['MJD']].values.copy()

# Perform clustering
clusterer = hdbscan.HDBSCAN(min_cluster_size=3)
df['Cluster'] = clusterer.fit_predict(X)

# Step 3: Now you can use the Cluster column
# Remove noise (points labeled as -1)
valid_df = df[df['Cluster'] != -1]

# Compute cluster size ΣΔν
cluster_sizes = valid_df.groupby('Cluster')['Δf (Hz)'].sum()

print("Cluster column created successfully!")
print("\nCluster distribution:")
print(df['Cluster'].value_counts())
print("\nCluster sizes (ΣΔf):")
print(cluster_sizes)
```

```python
Reservoir Model
```

```python

from scipy.optimize import curve_fit

# Load the data
df = pd.read_excel('\\Users\\hp\\Desktop\\MALACHY.xlsx')

# Use MJD directly instead of Waiting_Time
def reservoir_model(n, initial_mjd, slope):
    return initial_mjd + slope * n

n = np.arange(len(df))
waiting = df['MJD'].values  # Use MJD column instead

params, _ = curve_fit(reservoir_model, n, waiting)

print("Fitted params (initial_mjd, slope):", params)
```

```python
from sklearn.metrics import mean_absolute_error

predictions = reservoir_model(n, *params)

mae = mean_absolute_error(waiting, predictions)

print("MAE:", mae)
```
